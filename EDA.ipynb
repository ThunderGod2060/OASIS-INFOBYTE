```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('customer_shopping_data.csv', encoding= 'unicode_escape')
```

```python
EXPLORATORY DATA ANALYSIS
```

```python
PAYMENT METHOD
```

```python
ax = sns.countplot(x = 'payment_method' , data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
FROM THE ABOVE GRAPH WE CAN CONCLUDE THAT MOST TRANSACTIONS HAVE BEEN MADE THROUGH CASH, CREDIT CARD AND DEBIT CARD
RESPECTIVELY.
```

```python
GENDER 
```

```python
payment_gen = df.groupby(['gender'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)

sns.barplot(x='gender', y='price', data = payment_gen)
```

```python
From above graph be can conclude that female are the majority buyers and even spend more money then males.
```

```python
df.columns
```

```python
QUANTITY
```

```python
ax =sns.countplot(x = 'quantity' , hue = 'gender', data = df) 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
FROM THE ABOVE GRAPH WE CAN CONCLUDE THAT FEMALES HAVE PURCHASED MORE QUANTITY OF GOODS
```

```python
QUANTITY
```

```python
quantity_cat = df.groupby(['category'], as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

sns.barplot(x='quantity', y='category', data = quantity_cat)
```

```python
FROM THE ABOVE GRAPH WE CAN CONCLUDE THAT THE CATEGORY OF CLOTHING SOLD THE MOST
```
